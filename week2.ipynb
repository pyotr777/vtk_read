```python
import csv
import numpy as np
filename = "GLB.Ts+dSST.csv"
usecols=range(1,12)
separator=","
f = open(filename, 'r')
title = f.readline()
print("Title:" +title)
headers = f.readline().strip().split(separator)[1:13]
print headers
years_arr=[]
data_arr=[]
for line in f:
    raw_data = line.strip().split(separator)
    for i, r in enumerate(raw_data):
        if r[0]=="*":
            raw_data[i]="0"
    print(raw_data)
    years_arr.append(raw_data[:1])
    years = np.array(years_arr)
    data_arr.append(raw_data[1:13])
    
data = np.array(data_arr,dtype=np.float16)
print data.shape
print data
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
cmap = mpl.cm.brg
color_n = 1
n = len(headers)
plt.figure(figsize=(20,18))
for i in range(0,n):
    #print data[:,i]
    plt.plot(years,data[:,i],color=cmap(i / float(n)))
    
plt.legend(headers, loc='center left', bbox_to_anchor=(1, 0.5))
plt.title(title)
plt.show()
```
